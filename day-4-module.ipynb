```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20220309'

resp = request.urlopen(url)
soup = BeautifulSoup(resp)
print(soup)
```

```python
movie_title = []
movie_point = []

for line in soup.findAll('tr'):
#     print(line)
    title = line.find('div',class_='tit5')
    if title :
        movie_title.append(title.get_text().strip('\n'))
        
    point = line.find('td',class_='point')
    if point :
        movie_point.append(point.get_text())
        
for i, (title,point) in enumerate(zip(movie_title,movie_point)):
    print("{} {:<35}\t : {}".format(i+1,title,point))
```

```python
from wcwidth import wcswidth
def fmt(x, b=' ', w=0, align='r'):
    """ 동아시아문자 폭을 고려하여,
    문자열 포매팅을 해 주는 함수.
    w 는 해당 문자열과 스페이스문자가 차지하는 너비.
    align 은 문자열의 수평방향 정렬 좌/우/중간. """
    x = str(x)
    l = wcswidth(x)
    s = w-l
    if s <= 0:
        return x
    if align == 'l':
        return x + b*s
    if align == 'c':
        sl = s//2
        sr = s - sl
        return b*sl + x + b*sr
    return b*s + x

for i, (title,point) in enumerate(zip(movie_title,movie_point)):
    print("{} {} {}".format(fmt(i+1,w=2),fmt(title,'-',35,'l'),fmt(point)))
```

```python
from urllib import request
from bs4 import BeautifulSoup

url = 'https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'
target = request.urlopen(url)

soup = BeautifulSoup(target)

for location in soup.select('location'):
    print('도시\t:',location.select_one('city').string)
    print('날씨\t:',location.select_one('wf').string)
    print('최저기온:',location.select_one('tmn').string,'℃')
    print('최고기온:',location.select_one('tmx').string,'℃')
    print()
```
