```python
# 아이뉴스 24
from urllib import request
from bs4 import BeautifulSoup

url = 'https://www.inews24.com/list/020400'
domain = 'https://www.inews24.com'
target = request.urlopen(url)
soup = BeautifulSoup(target)
```

```python
def ptag_del(select):
    return ''.join(''.join(str(select).split('<p>')).split('</p>'))
```

```python
news_title = []
news_desc = []
news_datetime = []

for line in soup.findAll('li',class_='list'):
    href = line.find('a').get('href')
#     print(domain+href)
    new_target = request.urlopen(domain+href)
    news_one_soup = BeautifulSoup(new_target)
    p_tag = news_one_soup.find('article').select('p')
    for desc in p_tag:
        print(ptag_del(desc))
        print()
```
